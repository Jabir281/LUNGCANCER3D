# Download Trained Model
This notebook helps you verify and download your trained model (`best_model.pth`) from the Jupyter environment.

```python
import os
from IPython.display import FileLink
```

```python
# Define the model filename
model_path = 'best_model.pth'

# Check if file exists and print size
if os.path.exists(model_path):
    size_mb = os.path.getsize(model_path) / (1024 * 1024)
    print(f"Found model file: {model_path}")
    print(f"Size: {size_mb:.2f} MB")
else:
    print(f"Error: File {model_path} not found. Did the training finish?")
```

```python
# Generate download link
if os.path.exists(model_path):
    display(FileLink(model_path))
```
